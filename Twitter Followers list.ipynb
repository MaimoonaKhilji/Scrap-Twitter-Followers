```python
from twython import Twython
import csv

APP_KEY= ''
APP_SECRET= ''
OAUTH_TOKEN= ''
OAUTH_TOKEN_SECRET= ''


twitter = Twython(APP_KEY, APP_SECRET,
                  OAUTH_TOKEN, OAUTH_TOKEN_SECRET)
                  
followers = twitter.get_followers_list(screen_name = "geeksforgeeks",count=200)
following = twitter.get_friends_ids(screen_name = "geeksforgeeks")
with open("twitter.csv", 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerow(["name", "tweets", "following", "followers", "I'm following"])
    for e in followers["users"]:
        writer.writerow([e["screen_name"], e["statuses_count"], e["friends_count"], e["followers_count"], e["id"] in following["ids"]])
```
